```python
with open('emoji/emoji_final_join.json',encoding='utf-8') as f: 
    emoji_list = json.load(f)


def check_emoji(text):
    all_emoji = []
    polarity = 0
    
    for ele in emoji_list:
        if ele['emoji'] in text:
            all_emoji.append(ele['emoji'])
            polarity += ele['polarity']
    emojis = ''.join(all_emoji)
    
    if len(all_emoji) != 0:
        polarity = round(polarity/len(all_emoji))
        
    return emojis, polarity
            


for element in tweet_obj_list:
    emojis,polarity = check_emoji(element[0]['text'])
    element[0]['emoji'] = emojis
    element[0]['emoji_polarity'] = polarity
      
```
