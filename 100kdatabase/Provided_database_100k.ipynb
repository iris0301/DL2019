```python
import pandas as pd
import csv
```

```python
data = pd.read_csv('training.csv')
```

```python
negative = data.loc[data.Sentiment == 0,:]
negative = negative.loc[:,('Text', 'Sentiment')][:30000]
negative
```

```python
positive = data.loc[data.Sentiment == 4,:]
positive = positive.loc[:,('Text', 'Sentiment')][:30000]
positive
```

```python
data2 = pd.read_csv('largedata.csv')
```

```python
neutral = data2.loc[data2.Emoji.isnull(),:]
neutral = neutral.loc[neutral.Sentiment == 0,:]
neutral = neutral.loc[neutral.Sentiment140 == 2,:]
```

```python
neutral = neutral[:40000]
```

```python
neutral = neutral.loc[:,('Text', 'Sentiment140')]
neutral.rename(columns={'Sentiment140':'Sentiment'}, inplace=True)
neutral
```

```python
tweets_100k = pd.concat([positive,neutral,negative], ignore_index=True)
```

```python
tweets_100k['sentiment'] = tweets_100k.Sentiment.div(2).astype('int32')
tweets_100k = tweets_100k.loc[:,('Text', 'sentiment')]
tweets_100k.rename(columns={'sentiment':'Sentiment'}, inplace=True)
tweets_100k
```

```python
from sklearn.utils import shuffle
tweets_100k = shuffle(tweets_100k)
```
