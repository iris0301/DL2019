```python
import json
import os 
import urllib 
import re 
import requests 


################################### helper functions

def unicode(str):
    result = []
    split = str.split("-")
    for s in split:
        utf = "U+" + s
        result.append(utf)
    
    return result

################################### helper functions

data = []

with open('emoji.json') as f:
    raw = json.load(f)

# to access, add encoding
with open('emoji-descriptors.json', encoding='utf-8') as f:
    emoji_list = json.load(f)
    
with open('emoji_emotion.json', encoding='utf-8') as f:
    emoji_emotion = json.load(f)
    
    
#######################################

#unicodes is a string of unicodes without the u+
def join_tables(unicodes):
    unicode_split = unicodes.split("-")
    set_unicode = set(unicode_split)
    
    for l in emoji_list:
        # all the codes of emoji-descriptors in one array
        codes_split = l['codes'].split(" ")
        set_codes = set(codes_split)
        
        # we want to see if all of unicodes_split in code_slit
        
        if set_unicode == set_codes:
            return l
        
        
def check_None_emoji(obj):
    if obj == None:
        return None
    else:
        return obj['char']
    
def check_None_name(obj):
    if obj == None:
        return None
    else:
        return obj['short_name']

#######################################

    
for line in raw:
    uni = unicode(line['unified'])
    
    descriptor_line = join_tables(line['unified'])
    
    if descriptor_line!= None:
        emoji_icon = check_None_emoji(descriptor_line)
        data.append({
                'emoji': emoji_icon,
                'name': check_None_name(line),
                'unified': line['unified'],
                'unicode': uni,
                'category': line['category'],
                'has_img_twitter': line['has_img_twitter'],
                'polarity': 0
            })
    

for line in data:
    #print(line['name'])
     for l in emoji_emotion:
            if line['name'] == l['name']:
                line['polarity'] = l['polarity']

print(data)
                
                
        
with open('emoji_final_join.json', 'w', encoding='utf-8') as outfile:  
    json.dump(data, outfile, ensure_ascii=False, indent=4, separators=(',', ': '))
```
