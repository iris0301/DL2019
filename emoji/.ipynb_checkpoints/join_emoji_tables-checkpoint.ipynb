```python
import json
import os 
import urllib 
import re 
import requests 


################################### helper functions

def unicode(str):
    result = []
    split = str.split("-")
    for s in split:
        utf = "U+" + s
        result.append(utf)
    
    return result

################################### helper functions

data = {} 
data['emoji'] = []

with open('og/emoji.json') as f:
    raw = json.load(f)

# to access, add encoding
with open('og/emoji-descriptors.json', encoding='utf-8') as f:
    emoji_list = json.load(f)
    
    
#######################################

#unicodes is a string of unicodes without the u+
def join_tables(unicodes):
    unicode_split = unicodes.split("-")
    set_unicode = set(unicode_split)
    
    for l in emoji_list:
        # all the codes of emoji-descriptors in one array
        codes_split = l['codes'].split(" ")
        set_codes = set(codes_split)
        
        # we want to see if all of unicodes_split in code_slit
        
        if set_unicode == set_codes:
            return l
        
        
def check_None_emoji(obj):
    if obj == None:
        return None
    else:
        return obj['char']
    
def check_None_keywords(obj):
    if obj == None:
        return None
    else:
        return obj['keywords']
    
def check_None_name(obj):
    if obj == None:
        return None
    else:
        return obj['name']

def get_polarity(descriptor):
    if check_None_emoji(descriptor) == None:
        return 0
    else:
        return
        

    
#######################################

    
for line in raw:
    uni = unicode(line['unified'])
    
    descriptor_line = join_tables(line['unified'])
    
    if descriptor_line!= None:
        emoji_icon = check_None_emoji(descriptor_line)
        
        data['emoji'].append({
            'emoji': emoji_icon,
            'keywords': check_None_keywords(descriptor_line),
            'name': check_None_name(descriptor_line),
            'unified': line['unified'],
            'unicode': uni,
            'category': line['category'],
            'has_img_apple': line['has_img_apple'],
            'has_img_google': line['has_img_google'],
            'has_img_twitter': line['has_img_twitter'],
            'has_img_facebook': line['has_img_facebook'],
            'polarity:': get_polarity(descriptor_line)

        })

print(data)
    
    
with open('emoji_final_join.json', 'w', encoding='utf-8') as outfile:  
    json.dump(data, outfile, ensure_ascii=False, indent=4, separators=(',', ': '))
```

```python
import json
import os 
import urllib 
import re 
import requests 
from datetime import timedelta, date 
from time import sleep 
from bs4 import BeautifulSoup
import sqlite3
from pprint import pprint


with open('pretty_emoji.json', encoding='utf-8') as f:
    raw = json.load(f)


with open('og/emoji_polarity.json', encoding='utf-8') as f:
    polarity = json.load(f)


################## helper functions

def match_polarity(emoji):
    if emoji != None:
        for p in polarity:
            if p['emoji'] == emoji:
                return p['polarity']
    return 0


###################################
raw = raw['emoji']

# trial = raw[975]
# print(trial)
# polarity = match_polarity(trial['emoji'])
# print(polarity)

new_data = {} 
new_data['emoji'] = []

for r in raw:
#     print(r)
#     print("--")

    pa = match_polarity(r['emoji'])
    if pa!= None:
        new_data['emoji'].append({
            'emoji': r['emoji'],
            'keywords':r['keywords'],
            'name': r['name'],
            'unified': r['unified'],
            'unicode': r['unicode'],
            'category': r['category'],
            'has_img_apple': r['has_img_apple'],
            'has_img_google': r['has_img_google'],
            'has_img_twitter': r['has_img_twitter'],
            'has_img_facebook': r['has_img_facebook'],
            'polarity' : pa
        })

    

print(raw)
                         
with open('pretty_emoji.json', 'w', encoding='utf-8') as outfile:  
    json.dump(new_data, outfile, ensure_ascii=False, indent=4, separators=(',', ': '))
```
