```python
import json
import os 
import urllib 
import re 
import requests 
import tweepy           
import datetime
from time import gmtime, strftime
from textblob import TextBlob 
        
     
consumerKey =  "2XaGuf5hIS0K5l72OoYxLzklz"
consumerSecret = "LeJVqoOgOwLVE2AedE1WiFmEGG6ps1fhX6XMgm3qVBCyTO231u"
accessToken = "3328579866-cxEFxLWIz7KiHB9vvTNhplZlMOK5hbBBsSnfoNl"
accessTokenSecret = "xrY0e5mF6nk3WLamxkBlBxabb42PYrgJ3kurkEEcuvgxl"
  
   
auth = tweepy.OAuthHandler(consumerKey, consumerSecret)  
auth.set_access_token(accessToken, accessTokenSecret)  
  

api = tweepy.API(auth, wait_on_rate_limit=True)  


i=0
try:
    tweet_obj_list=[]
    for tweet in tweepy.Cursor(api.search,q = 'meal',tweet_mode='extended').items(50):    
        tweet_obj = []
        check_retweet = 1
        i = i+1
        print(i)
        a = tweet.full_text
        b = a[:4]
        check_retweet = int(b == 'RT @')
        #print(check_retweet)
                
        if check_retweet == 0 :   
            content = tweet.full_text
            content = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", content).split()) 
            analysis = TextBlob(content) 
                # set sentiment 
            if analysis.sentiment.polarity > 0: 
                sentiment = 'positive'
            elif analysis.sentiment.polarity == 0: 
                sentiment = 'neutral'
            else: 
                sentiment = 'negative'
            
            
            tweet_obj.append({
                'id': tweet.id,
                'text': tweet.full_text,
                'sentiment': sentiment,
                'user_id': tweet.user.id,
                'name': tweet.user.screen_name,
                'location': tweet.user.location,
                'retweet count': tweet.retweet_count,
                'favorite count': tweet.favorite_count,
                'time': tweet.created_at.strftime('%m/%d/%Y %H:%M:%S'),
                
                })
            print(tweet_obj)
            tweet_obj_list.append(tweet_obj)
        else:
            print('retweet')
except:
    print('error on: ')
    
    
## writes to this  
with open('meal-01.json', 'w', encoding='utf-8') as outfile:   
    json.dump(tweet_obj_list, outfile, ensure_ascii=False, indent=4, separators=(',', ': ')) 
    
```
